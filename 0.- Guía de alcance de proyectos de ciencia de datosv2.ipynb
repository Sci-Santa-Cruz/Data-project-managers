# Criterios Iniciales de Selección
Aquí asumiremos que ya hemos revisado los criterios básicos para realizar proyectos de ciencia de datos para el bien social:

- **Impactante:** El problema que estamos resolviendo es real, importante y tiene un impacto social.
- **Solucionable:** Los datos pueden desempeñar un papel en la resolución del problema, y la organización tiene acceso a los datos adecuados (consulte nuestro [marco de madurez de datos](enlace-a-tu-marco-de-madurez-de-datos) para ayudar a evaluar si tiene los datos adecuados).
- **Acción:** La organización ha priorizado este problema, está lista para tomar acciones basadas en el trabajo y está dispuesta a comprometer recursos para validar e implementar el proyecto (consulte nuestro [marco de madurez organizativa](enlace-a-tu-marco-de-madurez-organizativa) para evaluar su estado).

## Resumen del Alcance
Una vez que hemos pasado por el proceso de selección inicial, comenzamos el proceso de delimitación. Hay muchas formas de delimitar un problema. Nos centramos en proyectos que sean accionables y que conduzcan a resultados sociales positivos tangibles. El enfoque que describimos a continuación es uno que ha funcionado para nosotros, pero ciertamente no creemos que sea la única manera de delimitar un proyecto exitoso. Háganos saber si tiene enfoques alternativos que pueda ser beneficioso. Como siempre, el proceso de delimitación es bastante iterativo y el alcance se refina tanto durante el proceso de delimitación como durante el proyecto.

### ¿Quién debe participar?
Los participantes ideales en el proceso de delimitación incluyen a varias partes interesadas en el proyecto, incluidas las personas que comprenden el problema comercial/político, las personas que comprenden qué datos están disponibles, las personas que consumirán los resultados del sistema o tomarán medidas en función de él, y las personas que se verán afectadas por él.

### Paso 0: Comprensión del Problema
¿Cuál es el problema? ¿A quién afecta y en qué medida? ¿Cómo se está resolviendo actualmente y cuáles son algunas de las brechas?

### Paso 1: Objetivos
¿Cuáles son los objetivos del proyecto? ¿Cómo sabremos si nuestro proyecto es exitoso?

### Paso 2: Acciones
¿Qué acciones o intervenciones informará este trabajo?

### Paso 3: Datos
¿Qué datos tiene acceso internamente? ¿Qué datos necesita? ¿Qué se puede agregar de fuentes externas y/o públicas?

### Paso 4: Análisis
¿Qué análisis debe realizarse? ¿Implica descripción, detección, predicción o cambio de comportamiento? ¿Cómo se validará el análisis?

### Consideraciones Éticas
¿Cuáles son los problemas de privacidad, transparencia, discriminación/equidad y responsabilidad en torno a este proyecto y cómo los abordará?

### Consideraciones Adicionales
¿Cómo implementará su análisis como un nuevo sistema para que pueda actualizarse e integrarse en las operaciones de la organización? ¿Cómo evaluará el nuevo sistema en el campo para asegurarse de que cumpla con sus objetivos? ¿Cómo supervisará su sistema para asegurarse de que continúe funcionando bien con el tiempo?

El proceso de delimitación es iterativo y no estrictamente lineal. Si bien comprender el problema y definir los objetivos del proyecto debe ocurrir antes de identificar las acciones, el proceso de identificación de acciones puede ayudar a discernir si un objetivo es accionable y debe redefinirse. Del mismo modo, evaluar los datos disponibles para el proyecto puede llevarnos a replantear qué problemas, objetivos y acciones pueden ser informados por un proyecto de ciencia de datos. Nuestro análisis puede llevarnos a repensar nuestro problema y nuestros objetivos y comenzar de nuevo el proceso de delimitación. Por lo tanto, aunque cada paso sigue al anterior, debemos considerar cada paso como una oportunidad para evaluar los pasos anteriores.

En todo momento, la ética debe ser el centro de nuestro proceso de delimitación. Las cuestiones éticas deben considerarse en cada fase de la delimitación y ejecución del proyecto, en lugar de ser consideradas como un "paso" discreto en ese proceso. Desarrollar una comprensión clara y explícita de sus objetivos en este sentido es fundamental para hacerlo bien. Crucialmente, la integración de consideraciones éticas en el proyecto no debe ser una reflexión posterior ni una carga, sino más bien un área crítica y continua de enfoque que involucre a todas las partes interesadas, especialmente a las personas que se verán afectadas por este sistema.


# Paso 0: Comprender el problema
Antes de comenzar a delimitar el proyecto, debemos asegurarnos de comprender el problema y su impacto. El problema debe ser una prioridad para la organización y debe poder abordarse utilizando los datos que la organización tiene o puede acceder. Si el problema no es una prioridad, ni siquiera un modelo bien diseñado ayudará a resolverlo, ya que la organización carecerá de la motivación para actuar sobre la información resultante del análisis.

Para comenzar, es importante comprender el alcance del problema. Primero pedimos a las organizaciones que describan el problema que enfrentan, incluyendo quiénes o qué se ven afectados por el problema, cuántos se ven afectados y en qué medida (es decir, la magnitud del problema). Por ejemplo, un distrito escolar puede estar preocupado por las bajas tasas de graduación de la escuela secundaria. Deben poder decirnos quiénes se ven afectados, quizás estudiantes de bajos ingresos o estudiantes en riesgo. También deben poder decirnos la magnitud del problema: por ejemplo, que menos del 90% de los estudiantes de secundaria se gradúan a tiempo y menos de la mitad de los estudiantes en riesgo.

Luego, pedimos a la organización que explique por qué el problema es una prioridad en este momento y cómo han estado abordando el problema. Comprender cómo la organización ha abordado el problema puede ayudarnos a identificar formas en que el análisis de datos puede informar las acciones de la organización para lograr sus objetivos. Por ejemplo, un distrito escolar puede estar preocupado por las tasas de graduación de la escuela secundaria porque recientemente descubrieron que eran particularmente bajas entre los estudiantes en riesgo. El distrito escolar puede estar inscribiendo a los estudiantes en riesgo en programas de tutoría después de la escuela que ayudan a reforzar el aprendizaje en la escuela. Si el espacio en los programas después de la escuela es limitado, entonces el análisis adecuado podría ayudar al distrito escolar a priorizar a los estudiantes para la inscripción que es poco probable que se gradúen a tiempo.

Finalmente, es importante identificar los grupos o partes interesadas dentro y fuera de su organización que deben participar en la delimitación e implementación del proyecto. Por lo general, los proyectos de ciencia de datos requieren la participación de partes interesadas dentro de su organización, como los responsables de políticas, gerentes, propietarios de datos, propietarios de infraestructura de TI y las personas que intervendrán, como trabajadores de la salud. Los proyectos también suelen requerir la participación de personas y grupos fuera de su organización, como grupos comunitarios que se verán afectados por este trabajo. Por ejemplo, un distrito escolar que delimita un proyecto de ciencia de datos para identificar a los estudiantes en riesgo de no graduarse de la escuela secundaria querrá involucrar a los responsables de políticas de alto nivel, así como a los trabajadores de TI y datos temprano en su planificación. También querrán involucrar a los administradores escolares, maestros y trabajadores de programas de tutoría para ayudar a comprender el problema y delimitar el proyecto. El distrito también querrá involucrar a la junta escolar y a los padres de estudiantes que probablemente se verán afectados por el proyecto, especialmente en comunidades más afectadas por las bajas tasas de graduación.

Una vez que el problema se comprende correctamente, podemos comenzar a definir los objetivos de la organización.

## Paso 1: Definir el(los) objetivo(s)
El objetivo aquí es tomar el resultado que estamos tratando de lograr y convertirlo en un objetivo que sea medible y alcanzable. Este es el paso más crítico en el proceso de delimitación. En el contexto de nuestros proyectos, un objetivo es un objetivo o resultado concreto, específico y medible que la organización logrará abordando el problema. A menudo nos encontramos con esfuerzos en los que una organización definirá el objetivo de su proyecto como la construcción de una solución técnica, como un modelo predictivo, un panel de control o un mapa. Sostenemos que la solución técnica (modelo, panel de control, mapa) no es en sí misma el objetivo de un proyecto de ciencia de datos. Más bien, el objetivo del proyecto debería ser resolver algún problema de políticas u operaciones que afecta la misión de la organización. La construcción de una solución de ciencia de datos puede (y si se hace correctamente, debería) ayudarlo a alcanzar el objetivo que está buscando.

La mayoría de los proyectos comienzan con un objetivo muy vago y abstracto (por ejemplo, mejorar la educación), se vuelven un poco más concretos (aumentar el porcentaje de estudiantes que se graduarán a tiempo) y continúan refinándose hasta que el objetivo sea concreto, inequívoco y alcance los objetivos sociales de la organización. Por ejemplo, "mejorar la educación" es demasiado abstracto y vago para ser el objetivo de un proyecto específico de ciencia de datos, mientras que "aumentar el número de estudiantes que se graduarán de la escuela secundaria a tiempo" es un objetivo más apropiado. Este paso a menudo es difícil porque muchas organizaciones no han definido explícitamente objetivos concretos y analíticos para muchos de los problemas que están abordando. A veces, estos objetivos existen pero están implícitos en las mentes de las personas dentro de la organización. Otras veces, hay varios objetivos que diferentes partes de la organización están tratando de lograr.

Un enfoque que encontramos útil para definir objetivos es relacionarlo directamente con el problema que hemos identificado y, por lo general, mejorar/maximizar/aumentar o disminuir/mitigar/reducir un resultado o métrica relevante (por ejemplo, aumentar el porcentaje de estudiantes de secundaria que se gradúan a tiempo) que queremos cambiar.

A menudo, los objetivos también incluyen restricciones. Las restricciones suelen ser lo que hace que un proyecto de ciencia de datos sea necesario. Por ejemplo, si una agencia de salud pública pudiera inspeccionar todas las propiedades de alquiler en la ciudad en busca de violaciones del código de vivienda, probablemente lo harían. Sin embargo, puede haber una restricción en la cantidad de propiedades que una agencia puede inspeccionar en un período determinado, por lo que es posible que deseen priorizar aquellas que es más probable que sean inseguras para vivir. Un programa de salud pública preventiva podría querer que su objetivo sea minimizar la cantidad de visitas no planificadas a la sala de emergencias. Un enfoque trivial (y malévolo) para lograrlo sería cerrar todas las salas de emergencias, lo que resultaría en cero visitas. Agregar una restricción que requiera que la solución mejore los resultados de salud, y no solo reduzca las visitas a la sala de emergencias, nos ayudará a identificar soluciones que tengan el impacto social deseado.

Otras restricciones típicas relacionadas con objetivos son el presupuesto limitado, las personas y/o el tiempo limitado; restricciones legales o falta de voluntad política; o falta de licencia social.

Veamos algunos problemas de impacto social y cómo podemos definir los objetivos.

### Ejemplo 1: Envenenamiento por plomo
En 2014, trabajamos con el Departamento de Salud Pública de Chicago para reducir las tasas de envenenamiento por plomo en niños en Chicago. El objetivo inicial era reducir el envenenamiento por plomo aumentando la efectividad de sus inspecciones limitadas de peligros de plomo. Una forma de lograr ese objetivo sería enfocarse en las inspecciones en hogares que probablemente tengan peligros de plomo. Aunque es útil, este enfoque no lograría su objetivo real, que era evitar que los niños se envenenasen con plomo. Encontrar un hogar con peligros de plomo y remediarlo solo es beneficioso si hay una alta probabilidad de que haya un niño presente en el hogar (actualmente o en el futuro) que probablemente esté expuesto al plomo (y desarrolle envenenamiento por plomo). La siguiente iteración del objetivo fue aumentar el número de inspecciones que encuentren peligros de plomo en hogares donde hay un niño en riesgo (antes de que el niño se exponga al plomo). Eventualmente, llegamos al objetivo final: reducir la cantidad de niños que se envenenarán con plomo en el futuro debido a peligros de plomo en su residencia actual mediante 1) identificar qué niños tienen un alto riesgo de envenenamiento por plomo en el futuro y luego 2) dirigir las intervenciones a los hogares de esos niños para eliminar esos peligros de plomo.

### Ejemplo 2: Graduación a tiempo de la escuela secundaria
Uno de los desafíos que enfrentan las escuelas hoy en día es ayudar a sus estudiantes a graduarse a tiempo. Están interesados en identificar a los estudiantes que están en riesgo de no graduarse a tiempo y necesitan apoyo adicional. Cuando hablamos inicialmente con la mayoría de los distritos escolares, comienzan con un objetivo muy estrecho de predecir qué niños es poco probable que se gradúen a tiempo. El primer paso en nuestro proceso de delimitación es volver al objetivo de aumentar las tasas de graduación y preguntar si hay un subconjunto específico de estudiantes en riesgo que desean identificar.

¿Qué pasa si pudiéramos identificar a los estudiantes que tienen solo un 5% de probabilidad de no graduarse a tiempo en comparación con aquellos que tienen un 95% de probabilidad de no graduarse sin apoyo adicional? Si el objetivo es simplemente aumentar las tasas de graduación, el primer grupo es (probablemente) más fácil de intervenir e influir, mientras que el segundo grupo puede ser más desafiante debido a los recursos que necesitan. ¿El objetivo es maximizar la probabilidad promedio de graduación, o el objetivo es enfocarse en los niños más en riesgo y aumentar la probabilidad de que el 10% inferior de los estudiantes se gradúe? ¿O es el objetivo crear más equidad y reducir la diferencia en las tasas de graduación a tiempo entre aquellos que son más propensos a graduarse y aquellos que son menos propensos?

Todos estos son objetivos razonables, pero las escuelas deben comprender, evaluar y decidir cuáles son los objetivos más importantes para ellas. Esta conversación a menudo los hace pensar más profundamente en la definición de cuáles son los objetivos organizativos y los intercambios entre ellos. Un objetivo razonable con el que podemos terminar después del proceso de delimitación es minimizar las disparidades en las tasas de graduación entre diferentes grupos raciales mientras maximizamos las tasas de graduación en general.

### Ejemplo 3: Inspecciones
Hemos trabajado en varios proyectos que involucraban inspecciones. Algunos ejemplos incluyen:

- La Agencia de Protección Ambiental de EE. UU. y el Departamento de Conservación Ambiental del Estado de Nueva York, priorizando qué instalaciones inspeccionar para violaciones de eliminación de residuos.
- La Ciudad de Cincinnati, identificando propiedades en riesgo de violaciones del código.
- El Grupo del Banco Mundial, priorizando quejas de fraude y colusión para investigar.

En la mayoría de los problemas de inspección, hay muchas más entidades (hogares, edificios, instalaciones, negocios, contratos) para inspeccionar de las que hay recursos disponibles para llevar a cabo esas inspecciones.

El objetivo con el que la mayoría de las organizaciones comienzan es enfocar sus inspecciones en entidades que probablemente estén en violación de las regulaciones existentes. Si bien este es un buen comienzo, la mayoría de estas organizaciones nunca pueden inspeccionar todo lo que podría no estar en conformidad. El objetivo que realmente desean lograr es la disuasión: reducir el número total de instalaciones que estarán en violación. Un proceso de inspección ideal resultaría en la reducción real de las violaciones, ya sea que se encuentren o no, lo que puede no ser lo mismo que un proceso de inspección que tiene como objetivo ser eficiente y aumentar la tasa de aciertos (% de inspecciones que resultan en violaciones).

### Ejemplo 4: Recolección de desechos planificada
Hace algunos años, trabajamos con Sanergy, una empresa social con sede en Kenia. Despliegan baños portátiles en asentamientos urbanos informales y uno de sus mayores costos es contratar personas para vaciar los baños recolectando desechos de cada uno de ellos. Dado que el uso de los baños varía, es ineficiente vaciar cada baño todos los días, como se hacía al comienzo del proyecto. Para que puedan crecer y mantener los costos bajos, necesitaban un enfoque más adaptativo que pueda "optimizar" el programa para vaciar los baños. El objetivo, en este caso, es asegurarse de que no se vacíe en exceso el baño cuando no esté lleno, pero tampoco se permita que se llene porque entonces no es utilizable. Esto se traduce en una formulación que maximiza la cantidad de veces que se vacían los baños cuando están lo más cerca posible del 100% de llenado sin llegar al 100%. Una formulación diferente del objetivo podría ser minimizar la cantidad de veces que una persona va a usar el baño (pero no puede) porque estaba lleno y no utilizable con las limitaciones de recursos del personal que tienen para vaciar los baños.

Considerar los intercambios al definir los objetivos
Al comenzar a definir y priorizar objetivos, a menudo en torno a la eficiencia, la eficacia y la equidad, la conversación conduce a intercambios.

Al tratar con estudiantes que pueden necesitar apoyo adicional para graduarse a tiempo, ¿qué es lo que más le importa: encontrar a cada estudiante que pueda necesitar ayuda (que puede no necesitar el apoyo y posiblemente ser ineficiente) o priorizar la eficiencia y enfocarse solo en los estudiantes de los que está extremadamente seguro necesitarán el apoyo adicional (y, por lo tanto, perder muchos estudiantes) o priorizar la equidad para asegurarse de que los estudiantes ya desfavorecidos que pueden no tener otros recursos disponibles para ellos no se queden atrás?

¿Preferiría inspeccionar más hogares sin encontrar peligros de plomo en ellos (lo que es ineficiente) o preferiría perder hogares con niños que finalmente se envenenarán con plomo?

Al despachar y ubicar vehículos de respuesta de emergencia, ¿quiere asegurarse de que pueda llegar a todas las emergencias posibles en 10 minutos o quiere asegurarse de que pueda llegar a emergencias críticas en 3 minutos y a las no críticas en 20 minutos?

¿Qué tipos de errores está más dispuesto a cometer? Esa es una pregunta crítica que un buen proceso de delimitación plantea y trata de responder según las prioridades de la organización.

En términos de ciencia de datos, ¿preferiría tener más falsos positivos o más falsos negativos? ¿Desea que estos falsos positivos o falsos negativos estén equilibrados en diferentes grupos raciales, de género, edad o socioeconómicos? Por supuesto, esta decisión depende del impacto y el costo de esos errores, que a menudo es difícil (y a veces incómodo) de cuantificar. Esta es una conversación sobre los valores sociales y políticos que desea que estén incorporados en sus operaciones, y los responsables de políticas deben decidir qué objetivos políticos desean lograr, qué recursos tienen y qué resultados desean priorizar. El trabajo de la ciencia de datos se utiliza luego para respaldar e implementar esos objetivos políticos. La ciencia de datos puede ayudar a explorar el impacto de esos objetivos y comprender mejor las implicaciones, pero en última instancia, es una decisión política decidir sobre qué objetivos optimizar y debe incluir a todas las partes interesadas afectadas por este sistema.



# Paso 2: ¿Qué acciones/intervenciones estás informando?

El trabajo que realizamos solo puede tener un impacto si es accionable. ¿Qué acciones puede tomar la organización para alcanzar estos objetivos? ¿Cómo podemos informar sobre esas acciones para que sean más efectivas? Estas acciones a menudo deben ser bastante concretas: inspecciones en el hogar, inscripción de un estudiante en uno de tres programas extracurriculares, correos electrónicos dirigidos para recaudación de fondos o defensa, despacho de un vehículo de emergencia o programación de la recolección de residuos. 

Un proyecto bien definido idealmente tiene un conjunto de acciones que la organización está tomando y que se pueden mejorar mediante la ciencia de datos. Por ejemplo, si un departamento de salud pública está inspeccionando propiedades en busca de plomo, la ciencia de datos puede ayudar a determinar qué hogares inspeccionar.

No es necesario limitarse a mejorar las acciones existentes. A menudo, terminamos creando un nuevo conjunto de acciones. Sin embargo, generalmente es una buena estrategia centrarse primero en informar sobre las acciones existentes en lugar de comenzar con acciones completamente nuevas con las que la organización no esté familiarizada.

Enumerar el conjunto de acciones que existen (o se pueden agregar fácilmente) permite que el proyecto sea accionable. Si el análisis que se realizará posteriormente no informa sobre una acción, generalmente no ayudará a la organización a alcanzar sus objetivos.

Volviendo a nuestro ejemplo de aumentar las tasas de graduación a tiempo para los estudiantes de secundaria. Las escuelas pueden tomar una variedad de acciones, que incluyen:

- Crear nuevos programas de apoyo.
- Inscribir a estudiantes en programas de apoyo existentes.
- Programar una reunión con un consejero de orientación o mentor.

Todas estas acciones varían en varias dimensiones y requieren información clave para ejecutarse.

- Crear nuevos programas de apoyo: necesitamos saber qué programas deben crearse, qué capacidad deben tener y quién debe inscribirse en esos programas.
- Inscribir a estudiantes en programas de apoyo existentes: cada programa tendrá ciertas limitaciones de capacidad, por lo que necesitamos saber qué estudiantes deben tener prioridad en cada programa.
- Programar una reunión con un consejero de orientación/mentor: esta es una intervención que requiere muchos recursos, por lo que necesitamos saber qué estudiantes deben tener prioridad para esta intervención.

**Desglose de acciones**

Al elegir una acción para informar, es importante tener en cuenta sus objetivos y pensar en diferentes aspectos de la acción. En primer lugar, debes crear una lista de acciones que una organización está tomando o podría tomar para lograr su objetivo. Luego, debes pensar en quién llevará a cabo esta acción y quién se verá afectado por ella. Por ejemplo, en el caso de las inspecciones de plomo, un inspector de plomo inspecciona casas en busca de evidencia de plomo que se pueda remediar para evitar que un niño se exponga. También debes considerar cómo la organización elige actualmente las prioridades y con qué frecuencia lo hace. Por ejemplo, ¿se selecciona la lista de propiedades a inspeccionar todos los días? ¿Cada semana? ¿Cuántas se seleccionan?

También debes considerar a través de qué canales se puede llevar a cabo una acción. El canal a través del cual se puede llevar a cabo una acción puede tener importantes implicaciones para las limitaciones de capacidad. Por ejemplo, generalmente se requieren menos recursos para enviar un correo electrónico, un mensaje de SMS o incluso una carta que para hacer una llamada telefónica en vivo o realizar un contacto en persona. Las limitaciones de capacidad pueden ayudarte a decidir cuán valioso puede ser un proyecto de ciencia de datos para priorizar tus acciones, así como cuántas acciones puede completar la organización a la vez.

Finalmente, siempre debes considerar las implicaciones éticas de las acciones que informarás con tu proyecto de ciencia de datos y si la priorización de la acción mediante un análisis de ciencia de datos podría exacerbar desigualdades u otros resultados indeseables.

# Paso 3: ¿Qué datos tienes y qué datos necesitas?

Notarás que hasta ahora, en el proceso de definición, no hemos hablado de datos en absoluto. Esto es intencional, ya que queremos que estos proyectos se centren en el problema y no en los datos. Sí, los datos son importantes y a todos nos encantan, pero comenzar con los datos a menudo conduce a un análisis que puede no ser accionable o relevante para los objetivos que deseamos alcanzar.

Una vez que comprendamos nuestro problema y hayamos determinado nuestros objetivos y acciones, el siguiente paso es averiguar qué datos tenemos acceso dentro y fuera de la organización que serán relevantes para resolver el problema, así como qué fuentes de datos podemos necesitar acceder.

En primer lugar, debes hacer una lista de las fuentes de datos disponibles dentro de la organización. Este es un proceso iterativo, ya que muchas organizaciones pueden no tener una lista completa de sus fuentes de datos. A veces, (si tienes suerte) los datos pueden estar en un almacén de datos central e integrado, pero incluso entonces puedes encontrar individuos o departamentos que tengan datos adicionales o versiones diferentes del almacén de datos. Consulta nuestro Marco de Madurez de Datos para evaluar la preparación de tu organización para un proyecto de ciencia de datos.

Es una buena práctica recopilar información detallada sobre cada fuente de datos. Por ejemplo, querrás conocer lo siguiente:

- Granularidad o detalle: el nivel al que se recopilan los datos. Por ejemplo, algunos datos pueden recopilarse sobre estudiantes individuales, otros sobre escuelas y otros sobre vecindarios o distritos escolares.
- Frecuencia: con qué frecuencia se actualizan los datos. Por ejemplo, los datos pueden actualizarse anualmente, mensualmente, semanalmente, diariamente o incluso de manera inmediata (en tiempo real). Algunos datos solo se actualizan de manera ad hoc, por ejemplo, cada vez que la organización recibe un conjunto de datos actualizado.
- Historia: cuánto tiempo abarcan los datos hacia atrás. Contar con más datos históricos mejorará el análisis.
- Identificadores: por lo general, querrás que los datos incluyan identificadores confiables y únicos que te permitan vincularlos con otras fuentes de datos, como números de Seguro Social, números de seguro, números de identificación de estudiantes

 o direcciones.
- Propietario: la organización, departamento, grupo o personas que controlan el acceso a los datos. Por ejemplo, los distritos escolares pueden ser propietarios de algunos datos de estudiantes, pero otros datos de estudiantes pueden ser propiedad de escuelas individuales o agencias locales de educación.
- Almacenamiento: cómo se almacenan los datos. Algunos datos se almacenan en bases de datos, mientras que otros se pueden almacenar en hojas de cálculo, PDF, almacenes de datos o incluso en archivos impresos.
- Ética: cuestiones éticas que pueden estar asociadas con el uso de las fuentes de datos. Por ejemplo, el uso de algunas fuentes de datos puede requerir el consentimiento de las personas cuyos datos se utilizan. También puede haber ciertos protocolos de seguridad que se requieran para acceder y utilizar los datos.


**Coincidencia de los Datos con las Acciones**

Antes de iniciar un proyecto de ciencia de datos, es importante determinar si los datos a los que tienes acceso coinciden con las acciones que necesitas informar. Por ejemplo, si las acciones que queremos tomar para alcanzar nuestros objetivos se centran en el nivel individual, es probable que necesites datos a nivel individual. Si las acciones deben decidirse una vez al día, entonces necesitas que tus datos se actualicen todos los días. Es importante que coincidan la granularidad, frecuencia y horizonte temporal de las acciones con la granularidad, frecuencia y horizonte temporal de los datos que tienes.

**Datos Externos y/o Públicos**

Una vez que hayas determinado qué datos necesitas y qué datos existen dentro de la organización, debes averiguar qué datos externos o públicos puedes obtener para llenar los vacíos. Cada dominio a menudo tiene fuentes de datos comúnmente utilizadas que puedes incorporar en tu proyecto. La Encuesta de la Comunidad Americana es un buen ejemplo de una fuente de datos que debes usar en la mayoría de los proyectos que involucran algún componente geográfico en Estados Unidos. Los portales de datos abiertos (a nivel federal, estatal y local) también tienen datos que se pueden usar para complementar tus datos internos. Datos de llamadas al 311, datos de llamadas al 911 y datos de incendios son ejemplos de algunas fuentes de datos locales comúnmente encontradas. También debes considerar fuentes de datos comerciales que puedes comprar para complementar tus datos internos. Ejemplos incluyen la compra de datos de organizaciones como Acxiom y Experian sobre comportamiento de compra o hábitos de compra de medios de Nielsen.

También puede haber otros datos que te gustaría incluir en tu análisis a los que actualmente no tengas acceso o que sean difíciles de obtener. Es importante pensar más allá de las fuentes de datos disponibles y considerar datos que podrían ser útiles para tu proyecto, incluso si actualmente no están disponibles. Identificar datos útiles pero no disponibles puede ayudarte a identificar posibles fuentes de datos que pueden mejorar tu análisis. También puede haber sustitutos para esos datos que existen. Esto puede incluir datos que son muy seguros o difíciles de acceder, como videos de CCTV, registros telefónicos o registros de ADN. También puede incluir datos que requerirían un esfuerzo adicional para recopilar, como resultados de encuestas, o datos que requerirían cambios en el sistema para recopilar, como datos que se actualizan con más frecuencia o se recopilan a un nivel de granularidad diferente.

Si bien es importante pensar a lo grande cuando se considera qué datos pueden mejorar tu análisis, también es importante tener en cuenta consideraciones éticas, especialmente al acceder a fuentes de datos sensibles. Las fuentes de datos externos que no son públicas pueden requerir acuerdos legales adicionales y requisitos para acceder.

# Paso 4: ¿Qué análisis se debe realizar?

Una vez que tengamos los objetivos, acciones y datos disponibles identificados, el paso final en el proceso de planificación es determinar los análisis que realizaremos para informar sobre las acciones identificadas, utilizando los datos que tenemos para lograr nuestros objetivos. Vale la pena reiterar que el análisis no es el objetivo de un proyecto de ciencia de datos, y es aconsejable dejar esta parte de lado hasta que los objetivos, acciones y datos disponibles estén claramente definidos.

Los análisis pueden utilizar métodos y herramientas de diferentes áreas: informática, aprendizaje automático, ciencia de datos, estadísticas y ciencias sociales. Una forma de pensar en el análisis realizado es dividirlo en cinco tipos:

**Descripción:** Principalmente enfocado en comprender eventos y comportamientos que han ocurrido en el pasado. Típicamente, todos los proyectos contienen un componente de análisis descriptivo para obtener una comprensión de los datos. Sin embargo, es raro que un análisis descriptivo sea el único componente de análisis de un proyecto.

Por ejemplo:

- Explorar datos sobre estudiantes que no se han graduado a tiempo en el pasado, revelando posibles "tipos" de estudiantes que sufren el resultado adverso en comparación con otros con respecto a factores altamente correlacionados.

**Detección:** Menos centrado en el pasado y más centrado en eventos en curso. Las tareas de detección a menudo implican detectar eventos y anomalías que están ocurriendo actualmente.

Por ejemplo:

- Utilizar datos de texto de proyectos de ley legislativos para identificar sus áreas temáticas.
- Detectar transacciones fraudulentas con tarjetas de crédito.

**Predicción:** Centrado en el futuro y en predecir comportamientos y eventos futuros.

Por ejemplo:

- Predecir la probabilidad de que un estudiante se gradúe de la escuela secundaria a tiempo en el momento en que ingresa al noveno grado.
- Predecir la probabilidad de que un proyecto de ley legislativo se convierta en ley en el momento en que se presenta en la legislatura.

**Optimización:** Centrado en tomar los resultados de otros tipos de análisis y utilizarlos para asignar recursos o tomar decisiones.

Por ejemplo:

- Identificar dónde ubicar ambulancias de manera que se maximice su cobertura.
- Dado un lista de estudiantes clasificados por su probabilidad de no graduarse a tiempo, identificar el subconjunto de estudiantes para intervención que tendría el mayor impacto.

**Cambio de comportamiento:** Centrado en causar un cambio en el comportamiento de las personas, organizaciones y vecindarios. Típicamente utiliza métodos de inferencia causal y economía conductual.

Por ejemplo:

- Dado que un estudiante corre el riesgo de no graduarse a tiempo y un conjunto de intervenciones posibles, identificar la intervención que maximizaría su probabilidad de graduarse a tiempo.

Por supuesto, hay muchos más tipos de análisis, pero nos centraremos en estos cinco aquí.

Para cada análisis que realicemos en el proyecto, debemos buscar responder lo siguiente:

- ¿Qué tipo de análisis se debe realizar y cuál es su propósito?
- ¿Es este un análisis descriptivo, un modelo predictivo o una tarea de detección o cambio de comportamiento? A menudo, el proyecto involucra varios de los tipos de análisis que describimos anteriormente, cada uno diseñado para informar acciones específicas y lograr objetivos específicos.

- ¿Cómo informará el análisis a las acciones?
- Cada análisis debe informar una o más de las acciones identificadas. Algunos análisis pueden informar múltiples acciones y a veces una acción puede ser informada por múltiples análisis.

- ¿Cómo se validará el análisis? ¿Qué validación se puede hacer utilizando datos históricos existentes?
- Es importante pensar en cómo podemos validar cada análisis utilizando los datos históricos disponibles. Deberíamos elegir la configuración de validación que refleje lo más cerca posible el escenario de implementación (cómo un usuario usaría este modelo, por ejemplo). Típicamente, para tareas de predicción, crearíamos múltiples conjuntos de validación y elegiríamos una métrica que refleje con precisión cómo usamos el modelo y observaríamos cómo funcionan nuestros modelos con el tiempo. Una cosa crítica que necesitaríamos pensar es el punto de referencia o "línea de base" con la que compararemos nuestros modelos. Si la organización está realizando esta tarea hoy de alguna manera, una buena línea de base es el rendimiento del "sistema existente". Baselines adicionales incluyen heurísticas simples (basadas en conocimiento experto o investigaciones previas). Queremos que nuestro nuevo análisis sea "mejor" que enfoques existentes o más simples para que valga la pena implementar el sistema de ciencia de datos, que es más caro de construir y mantener.

- ¿Cuáles son los problemas éticos asociados con el análisis?
- ¿Hay implicaciones de equidad que nos preocupan? ¿Cómo afectarían los errores en su análisis a individuos y a la sociedad?

Mostraremos esto con algunos ejemplos.

**Reducción de las tasas de envenenamiento por plomo**

**Objetivo:** Reducir la cantidad de niños que se envenenarán con plomo en el futuro debido a peligros de plomo en su residencia actual.

**Acciones:** Inspecciones preventivas en el hogar limitadas por recursos de inspección limitados.

**Análisis:** Necesitamos responder la pregunta clave: "¿Qué hogares debemos priorizar para inspecciones preventivas?" En otras palabras, queremos identificar los hogares de niños que corren el riesgo de envenenamiento con plomo en el futuro.

Dado que queremos intervenir antes de que un niño esté expuesto al plomo, predeciríamos la probabilidad de que cada niño menor de 2 años esté expuesto al plomo en sus hogares. Podríamos usar nuestras predicciones para clasificar los hogares en función del riesgo de que un niño esté expuesto al plomo en ellos y priorizarlos para la inspección. Si consideramos los cinco tipos de análisis que mencionamos anteriormente, esto sería una tarea de predicción.

**Validación:** Supongamos que la cantidad de hogares que la agencia puede inspeccionar en un mes es de 100. Dado esto, podríamos optimizar el análisis para predecir los 100 hogares donde un niño tiene más probabilidades de estar expuesto al plomo cada mes, una métrica que llamamos Precisión en K (o P@K).

En este ejemplo, el vínculo causal entre la acción (inspección y eliminación de plomo) y el objetivo (proteger a los niños de la exposición al plomo) está bien establecido. Eso podría no ser el caso en cada proyecto. Tomemos el ejemplo de la graduación oportuna de la escuela secundaria:

**Mejorar las tasas de graduación**

**Objetivo:** Aumentar el porcentaje de estudiantes de secundaria que se gradúan a tiempo, reduciendo al mismo tiempo la disparidad en las tasas de graduación entre grupos raciales.

**Acciones:** Proporcionar apoyo adicional a los estudiantes identificados como en riesgo de no graduarse a tiempo.

**Análisis:**

Al igual que con el proyecto de peligros de plomo, debemos realizar un análisis predictivo; predecir qué estudiantes tienen menos probabilidades de graduarse a tiempo. Podemos usar estas predicciones para priorizar un subconjunto de estudiantes para un alcance y apoyo adicionales.
Una vez que se predice el riesgo para cada estudiante, seleccionar el subconjunto de estudiantes para la intervención podría no ser tan sencillo como priorizar a los estudiantes con la puntuación de riesgo más alta. Dado la naturaleza de las intervenciones y la métrica que le interesa mejorar, los mejores estudiantes para la intervención podrían cambiar. Por ejemplo, si el objetivo es mejorar la tasa promedio de graduación con el mínimo esfuerzo, priorizar a los estudiantes con una puntuación de riesgo más intermedia podría resultar más efectivo que priorizar a aquellos que tienen muy poca probabilidad de graduarse a tiempo. Un análisis de optimización podría ayudar a identificar el conjunto óptimo de estudiantes a priorizar para la intervención.
Incluso si seleccionamos el conjunto "óptimo" de estudiantes para la intervención, podría haber múltiples tipos adicionales de apoyo disponibles (por ejemplo, tutoría después de clases, asesoramiento, ayuda con el transporte, ayuda financiera). Identificar la intervención más efectiva (s) para estudiantes individuales podría requerir un análisis de inferencia causal.


# Problemas Éticos

Como recordatorio, la consideración de problemas éticos debe estar integrada en cada fase de la planificación y ejecución de un proyecto, en lugar de ser considerada como un "paso" discreto en ese proceso. A menudo, la conversación inicial sobre ética en la fase de planificación se centrará en los valores éticos y sociales que deseamos incorporar en el sistema. Ten en cuenta que esta conversación no se limita a aplicaciones de ciencia de datos o inteligencia artificial, sino que se refiere a los valores que necesariamente deben incorporarse en los procesos de toma de decisiones que afectan la vida de las personas. Como tal, estas decisiones no pueden ser tomadas unilateralmente por los científicos de datos o tecnólogos encargados de construir el sistema. Aunque desempeñan un papel importante en la discusión, es importante que esta sea inclusiva de todas las partes interesadas relevantes: legisladores, tomadores de decisiones, desarrolladores de sistemas, propietarios de datos y la comunidad que se ve afectada por el sistema, todos tienen perspectivas importantes sobre estos temas.

A medida que avanzas en el proceso de planificación (y el proyecto en sí), ser explícito acerca de cómo tu trabajo refleja esos valores puede servir como un principio rector en el curso de las muchas decisiones que deben tomarse para llevar a cabo un proyecto. Para destacar su naturaleza central, hemos consolidado muchos de ellos aquí, planteando algunas preguntas motivadoras para ayudarte a explorar estos problemas. Esta lista está lejos de ser exhaustiva, pero proporciona un punto de partida para esa conversación:

## Privacidad, Confidencialidad y Seguridad

Las preocupaciones sobre la privacidad y la seguridad de los datos son comunes en aplicaciones de la ciencia de datos a problemas socialmente impactantes, especialmente cuando se involucran decisiones de alto riesgo, como la atención médica, la justicia penal y la educación. En algunos casos, puede haber requisitos legales que rigen cómo se pueden utilizar los datos y los pasos que deben estar en su lugar para protegerlos, pero las consideraciones éticas pueden ir más allá de lo que se requiere legalmente. Es especialmente importante aquí cómo las personas pueden sentirse acerca de los datos sobre ellas que se están utilizando, así como sus expectativas sobre la disponibilidad pública de esos datos. Preguntas a hacer aquí incluyen:

- ¿Cuáles son las consideraciones de privacidad (tanto legales como éticas)?
- ¿Cómo se protege la privacidad de las personas en los datos?
- ¿Qué hay de la confidencialidad?
- ¿Cuáles son las consideraciones y protecciones de seguridad? ¿Quién tiene acceso a qué partes de los datos? ¿Para qué fines? ¿Cuál es el proceso de auditoría de seguridad?

## Transparencia

Las consideraciones de transparencia para la construcción e implementación de sistemas de ciencia de datos pueden involucrar a muchas partes interesadas diferentes: actores y tomadores de decisiones internos, individuos cuyos datos se están utilizando, individuos que serán afectados por las decisiones que el sistema informa y el público en general. Al considerar las siguientes preguntas, ten en cuenta las perspectivas de cada una de estas partes interesadas:

- ¿Qué partes interesadas deben conocer sobre el proyecto?
- ¿Las personas cuyos datos estás utilizando saben que los estás utilizando?
- ¿Cómo afectarán las acciones que estás tomando basadas en este análisis a las personas? ¿Cuáles son los costos o beneficios de esta acción para las personas afectadas?
- ¿Las personas a las que estás priorizando saben si y por qué se les está priorizando?
- ¿Qué recurso tienen las personas afectadas por el nuevo sistema para impugnar o cambiar una decisión informada por tu análisis que las haya impactado?

## Discriminación, Equidad e Imparcialidad

Comprender y mejorar la imparcialidad de los sistemas de aprendizaje automático ha sido objeto de mucha escritura e investigación recientes, tanto en trabajos académicos como en la prensa popular. Durante el proceso de planificación, es importante comprender los tipos de disparidades que podrían resultar de tu proyecto y cómo podrían afectar a las personas, teniendo en cuenta las perspectivas de las personas impactadas por tu análisis y cualquier discriminación histórica y continua que pueda afectarles.

Por ejemplo, en el contexto de la detección de llamadas de emergencia relacionadas con el bienestar infantil, las disparidades en los falsos negativos (no realizar un seguimiento de un informe con una investigación cuando realmente debería hacerse) significarían más daño a los niños en una comunidad en comparación con otra. Por otro lado, las disparidades en los falsos positivos (realizar una investigación innecesaria) podrían resultar en una mayor presencia policial en algunas comunidades, lo que provocaría daños más amplios en la sociedad. Dos herramientas de nuestro trabajo que pueden ser útiles aquí son el "Árbol de Equidad", que utilizamos para facilitar estas conversaciones, y "Aequitas", que ayuda a auditar las salidas del modelo en busca de disparidades, ambas se pueden encontrar aquí. Algunas de las preguntas clave a hacer con respecto a la discriminación y la equidad son:

- ¿Qué grupos específicos podrían verse afectados de manera desigual por tu análisis, y cómo debes tener en cuenta esto?
- Desde la perspectiva de cada uno de estos grupos, ¿cómo definen la equidad en este contexto? ¿Hasta qué punto están alineadas estas definiciones?
- ¿Durante qué período de tiempo estás tratando de mejorar la equidad en este trabajo?
- ¿Cómo detectarás sesgos o desigualdades en tu sistema?
- ¿Qué disparidades existen en el proceso actual de toma de decisiones?
- Si existen inequidades, ¿cómo abordarás la reducción de ellas en tu sistema o la mitigación de su impacto en las decisiones e intervenciones posteriores?
- ¿Existen fuentes más amplias de inequidad, ya sean históricas o actuales, que afecten a los resultados que estás tratando de mejorar? ¿Cómo debe tu sistema tener en cuenta esto?

## Responsabilidad

La responsabilidad debe considerarse con respecto a todo el proceso, incluyendo las decisiones técnicas tomadas al trabajar con los datos y las decisiones tomadas sobre cómo se utilizará el sistema en la práctica y cómo se describirá al público. Asegurarse de que haya claridad desde el principio sobre quién es responsable y responsable de cada aspecto del sistema, especialmente en lo que respecta a posibles aspectos y consideraciones éticas, puede ayudar a

 reducir tanto los riesgos de que surjan problemas (al establecer límites y hacer que la supervisión sea responsabilidad explícita de alguien) como los posibles daños si surgen estos problemas (al tener planes de contingencia en su lugar). Por ejemplo, nuestro grupo ha trabajado en varios proyectos para reducir el riesgo de encarcelamiento en la cárcel mediante la provisión de intervenciones de apoyo, como apoyo de salud mental o servicios sociales. Sin embargo, el mismo modelo que predice el riesgo de futuros arrestos también podría utilizarse de manera más preocupante y punitiva, y nuestros acuerdos de colaboración deben definir cuidadosamente y proteger contra esto. Algunas preguntas a hacer sobre la responsabilidad incluyen:

- Si se utiliza datos sensibles para el proyecto, ¿quién es responsable de mantenerlos seguros? ¿Qué planes de contingencia existen en caso de una fuga y quién será responsable?
- ¿Qué limitaciones existen en la forma en que se utilizará tu análisis? ¿Quién es responsable si se utiliza para fines dañinos o no previstos?
- ¿Cómo supervisarás las consecuencias no deseadas (por ejemplo, si los estudiantes predichos en riesgo de bajo rendimiento interiorizan esta predicción)? ¿Qué procesos se implementarán para reducir cualquier daño potencial de ellos?
- ¿Cómo determinarás si el sistema está aumentando las disparidades con el tiempo? ¿Quién es responsable de supervisar esto y tomar decisiones sobre cómo responder?
- ¿Quién es responsable de determinar qué se debe divulgar sobre el sistema al público y cómo describirlo? Si las personas se oponen a ello, ¿quién es responsable y qué se debe hacer?
- ¿Qué recurso tienen las personas impactadas por el nuevo sistema si comete errores sobre ellas o hace recomendaciones basadas en datos inexactos? ¿Quién es responsable de responder a estos problemas y realizar las mejoras necesarias en el modelo o análisis subyacente?

## Licencia Social

El concepto de licencia social está relacionado con las ideas de transparencia y responsabilidad, pero con un enfoque diferente que puede proporcionar un ejercicio de reflexión útil. Independientemente de tus planes reales sobre la transparencia descrita anteriormente o cuántas personas pueden conocer los detalles del proyecto en la práctica, aquí debes pensar en cómo reaccionarían las personas a tu proyecto si recibiera una cobertura generalizada. ¿Sería uniformemente positivo o negativo, o en qué medida diferentes personas reaccionarían de manera diferente? Algunas preguntas para ayudarte a considerar la licencia social para tu proyecto incluyen:

- Si toda la población del país se entera de tu proyecto, ¿estarán de acuerdo con él? ¿Por qué?
- Si estuviera en la portada del periódico, ¿cuál sería el titular? ¿Sería positivo o negativo?
- Incluso si crees que la población podría apoyar tu proyecto en general, ¿hay grupos que podrían objetar? ¿Quiénes son y qué preocupaciones tendrían?

## Otras Consideraciones Éticas

Finalmente, es importante tener en cuenta que la lista aquí es solo un punto de partida y puede haber otras consideraciones éticas relevantes para tu contexto específico. ¿Existen requisitos legales, políticos u organizativos adicionales que no se mencionan anteriormente? ¿Necesitas obtener el consentimiento informado de las personas potencialmente afectadas por este trabajo? ¿Existen procedimientos específicos de supervisión, informe o revisión ética? Al igual que en los otros aspectos mencionados anteriormente, puede ser útil aquí pensar en el proyecto desde las perspectivas de diferentes partes interesadas y explorar si surgen consideraciones o preocupaciones que no encajan bien en las otras categorías aquí.

# Consideraciones Adicionales

## Evaluación

Una vez que tu análisis esté completo y lo hayas validado con datos disponibles, es importante evaluarlo en el campo para asegurarte de que ayude a alcanzar tus objetivos. Las pruebas en el campo son un tema complicado y pueden ser bastante específicas para el proyecto y sus objetivos, por lo que no entraremos en detalles aquí. Sin embargo, tu evaluación debe ser una aplicación del análisis a las acciones que se propuso informar y debe medirse según tus objetivos predefinidos.

## Implementación

Si el análisis resulta efectivo en la evaluación, querrás implementarlo como un nuevo sistema en la infraestructura de la organización para que se actualice regularmente e se incorpore en sus operaciones. Entre las consideraciones más importantes al implementar un nuevo sistema está si se construyó o no en la infraestructura de la organización. Si el sistema se construyó en la infraestructura de la organización, a menudo (aunque no siempre) será más fácil de implementar, porque debería estar en consonancia con la infraestructura técnica de la organización. Los científicos de datos que trabajan dentro de la organización generalmente tendrán acceso a la infraestructura interna de la organización y deberán trabajar en estrecha colaboración con los ingenieros de producción para asegurarse de que su sistema se integre con ella. Los voluntarios y otros científicos de datos externos también pueden acceder a la infraestructura de la organización a través de acuerdos voluntarios o comerciales, lo que facilita el proceso de implementación.

Sin embargo, a menudo los sistemas de ciencia de datos se construyen fuera de la infraestructura de la organización, posiblemente debido a problemas de seguridad o privacidad que limitan el acceso. Cuando esto sucede, la implementación puede ser más complicada. Los profesionales internos de datos e ingenieros de la organización que trabajan con y mantienen su infraestructura técnica deben incorporarse al proceso de planificación desde el principio para asegurarse de que el sistema construido por científicos de datos asociados se adapte a la infraestructura de la organización y se pueda implementar de manera bastante fácil. El proyecto debe ser una prioridad para los profesionales de datos y técnicos de la organización que son necesarios para la implementación, y debe haber tiempo reservado en sus horarios para dedicarse a este proyecto, al igual que con cualquier otro. Si los propios profesionales de datos y técnicos de la organización no están involucrados en el proyecto y no tienen tiempo presupuestado para el proyecto, entonces está destinado a fracasar. Los mismos profesionales de datos y técnicos que ayudan a implementar el nuevo sistema pueden necesitar mantenerlo en producción, y esto debe tenerse en cuenta en su horario y carga de trabajo.

Los sistemas de ciencia de datos también requieren recursos informáticos para la implementación. Antes de comenzar el proyecto, se debe tener una buena

 comprensión de la capacidad informática de la organización. En la implementación, los sistemas de ciencia de datos utilizarán memoria y llevarán tiempo cuando se actualicen. Esto debe tenerse en cuenta dentro de la infraestructura de la organización, y los profesionales de datos y técnicos internos deben asegurarse de que las actualizaciones regulares de los sistemas de ciencia de datos no entren en conflicto o compitan por recursos con otros procesos técnicos importantes.

Los sistemas de ciencia de datos también dependen de actualizaciones regulares y consistentes de las fuentes de datos subyacentes. Los profesionales de datos y técnicos de la organización deben asegurarse de que las actualizaciones regulares del nuevo sistema sean coherentes con las actualizaciones de los datos subyacentes. Si los datos son antiguos, están retrasados o de alguna manera no son consistentes con las actualizaciones del nuevo sistema, el valor de ese sistema se reducirá e incluso puede deteriorarse con el tiempo.

Finalmente, será importante considerar quién en la organización utilizará la salida del sistema y cómo accederán a ella. La clave para esto es pensar en cómo el usuario final utilizará realmente la información. ¿El usuario final estará revisando casos individuales uno por uno, o estará creando una lista para priorizar? Si el usuario final está considerando casos uno por uno, entonces una interfaz visual de usuario que proporcione la salida junto con otra información sobre el caso puede ser útil. Si, en cambio, el usuario final está creando una lista priorizada (por ejemplo, de lugares para inspeccionar), entonces puede ser mejor entregarle una lista de los casos principales, tal vez en una interfaz visual de usuario o incluso como un correo electrónico o documento imprimible. Estas decisiones deben tener en cuenta los procesos operativos actuales del usuario final, su acceso a la tecnología en su trabajo y su alfabetización (digital y de otro tipo), entre otras consideraciones. La capacitación también será una parte clave de la implementación, tanto para el mantenimiento técnico del sistema como para los usuarios finales que lo utilizarán para tomar decisiones.

## Monitoreo

Una vez que el sistema de ciencia de datos esté implementado e integrado en las operaciones de la organización, deberá monitorearse continuamente para asegurarse de que continúe funcionando bien y proporcione información que mejore la toma de decisiones. El rendimiento de los sistemas de ciencia de datos puede deteriorarse con el tiempo por diversas razones, incluidos cambios en los datos subyacentes o la calidad de los datos, cambios en las políticas u operaciones de la organización o incluso cambios en las tendencias del mundo real. Para garantizar que la información que proporciona el sistema siga siendo útil, las organizaciones deben tener un plan y asignar recursos para monitorear el rendimiento del sistema a lo largo del tiempo como parte regular de sus operaciones.

Primero, debes crear un plan sobre cómo se monitoreará el sistema, incluyendo qué datos y métricas se utilizarán para evaluar su rendimiento con el tiempo. Estos datos probablemente serán actualizaciones futuras de los mismos datos que se utilizaron para desarrollar el análisis original y pueden utilizar algunas de las mismas métricas y diagnósticos que se utilizaron para validarlo. Al monitorear el rendimiento, debes tener cuidado de asegurarte de que los datos utilizados para evaluar el sistema no se filtren en los datos utilizados para actualizarlo. Esto podría llevar a una sobreestimación del rendimiento del sistema y ocultar cualquier deterioro.

La organización también necesitará comprometer al personal técnico para monitorear el rendimiento del sistema y gestionarlo si se rompe o si el rendimiento se deteriora. Esto requiere comprender no solo el sistema, sino también los datos que se utilizan para actualizarlo y evaluarlo. Es posible que el personal técnico necesite capacitación para comprender, evaluar y actualizar el sistema, y se recomienda cierta redundancia para que la organización tenga la profundidad de habilidades para mantener el sistema, incluso si un miembro del personal se va.

La organización también debe planificar para llevar a cabo evaluaciones adicionales en el campo (como la descrita anteriormente) a intervalos regulares para asegurarse de que el sistema siga proporcionando información que sea utilizable y mejore las operaciones. Esto requiere el compromiso no solo de profesionales de datos, sino también de los usuarios finales en la evaluación del sistema. La frecuencia de estas evaluaciones puede variar según la capacidad de la organización, la frecuencia con la que se actualizan los datos y el sistema, y la frecuencia de los cambios en las operaciones de la organización y el entorno político general. Sin embargo, es importante que la organización planee dedicar recursos a la supervisión y evaluación continua del sistema. Después de todo, un sistema que proporciona información deficiente podría ser peor que no tener sistema alguno.

Finalmente, debes planificar qué hacer si el sistema muestra signos de deterioro. Este plan debe considerar cuáles pueden ser las fuentes de deterioro, incluidos cambios en los datos de entrada o en la evaluación de datos, cambios en las políticas u operaciones de la organización o cambios en las tendencias del mundo real. Las respuestas diferirán según las fuentes de deterioro del rendimiento.

También debes considerar qué hará la organización con el sistema si cambia su infraestructura, datos, políticas u operaciones. El sistema debe evaluarse bajo los nuevos cambios y debe determinarse si todavía proporciona información útil y utilizable o si deberá actualizarse o retirarse. Esto es importante porque los cambios en los datos de entrada o en la infraestructura que respalda el sistema podrían afectar su rendimiento, mientras que los cambios en la política u operaciones pueden significar que el sistema ya no proporciona información relevante para esos nuevos procesos.

# Resumen

Con suerte, esto proporciona a las agencias gubernamentales y organizaciones sin fines de lucro una visión general de cómo planificar proyectos de ciencia de datos/análisis y qué preguntas deben responder antes de lanzar el proyecto.

**Paso 1: Objetivos** – Define el objetivo o los objetivos del proyecto.

**Paso 2: Acciones** – ¿Qué acciones o intervenciones tienes que este proyecto informará?

**Paso 3: Datos** – ¿Qué datos tienes acceso internamente? ¿Qué datos necesitas? ¿Qué puedes obtener de fuentes externas o públicas?

**Paso 4: Análisis** – ¿Qué análisis necesita realizarse? ¿Involucra descripción, detección, predicción, optimización o cambio de comportamiento? ¿Cómo se validará el análisis?

**Consideraciones Éticas** – ¿Cómo

 abordarás las consideraciones éticas en cada fase del proyecto?

**Consideraciones Adicionales** – ¿Cómo se implementará y monitoreará el sistema de ciencia de datos una vez que esté en funcionamiento?

El objetivo de esta guía es ayudar a las organizaciones a planificar y ejecutar proyectos de ciencia de datos/análisis de manera efectiva y ética, para que puedan lograr sus objetivos de manera informada y con un impacto positivo en la sociedad.
